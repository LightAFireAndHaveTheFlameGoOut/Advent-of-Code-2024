```python
List = open("List.csv", "r")
Lists = List.read().split(",")
List1 = []
List2 = []
for i in range(len(Lists)):
    Lists[i] = int(Lists[i])
    if i % 2 == 0:
        List1.append(Lists[i])
    else:
        List2.append(Lists[i])
List.close()

List1 = str(List1).strip("[]")
List2 = str(List2).strip("[]")


List1file = open("List1.csv", "w")
List1file.write(List1)
List1file.close()

List2file = open("List2.csv", "w")
List2file.write(List2)
List2file.close()     
```

```python
List1 = open('List1.csv', 'r').read().split(',')
List2 = open('List2.csv', 'r').read().split(',')
List3 = []
fnum = 0

for i in range(len(List1)):
    List1[i] = int(List1[i])
for i in range(len(List2)):
    List2[i] = int(List2[i])
List1.sort()
List2.sort()

for i in range(len(List1)):
    if List1[i] > List2[i]:
        List3.append(List1[i] - List2[i])
    else:
        List3.append(List2[i] - List1[i])
        
for i in List3:
    fnum += i

print(fnum)
```
