```python
Core = open('BaseData.txt').read()
Core.replace(' ', 'x')
Count = 0
for i in range(len(Core)):
    if Core[i] != 'm' and Core[i] != 'u' and Core[i] != 'l' and Core[i] != '(' and Core[i] != ')' and Core[i] != ',' and not(Core[i].isdigit()) and Core[i] != 'd' and Core[i] != 'o' and Core[i] != 'n' and Core[i] != '\'' and Core[i] != 't':
        Core = Core.replace(Core[i], 'x')


for i in range(len(Core)-1):
    
    if (Core[i] == 'x' and (Core[i+1] != 'm' and Core[i+1] != 'd')) or (Core[i] == 'm' and Core[i+1] != 'u') or (Core[i] == 'u' and Core[i+1] != 'l') or (Core[i].isdigit() and (not(Core[i+1].isdigit()) and Core[i+1] != ',' and Core[i+1] != ')')) or (Core[i] == ')' and (Core[i+1] != 'm' and Core[i+1] != 'd')) or (Core[i] == 'd' and Core[i+1] != 'o') or (Core[i] == 'o' and (Core[i+1] != 'n' and Core[i+1] != '(')) or (Core[i] == 'n' and Core[i+1] != '\'') or (Core[i] == '\'' and Core[i+1] != 't') or (Core[i] == 't' and Core[i+1] != '('):
        #print(Core[i])
        Core = Core[:i+1] + 'x' + Core[i+2:]

    if (Core[i] == 'x' and (Core[i+1] == 'm' or Core[i+1] == 'd')) or Core[i] == ')':
        Count += 1

for i in range(len(Core) + Count - 1):
    if (Core[i] == 'x' and (Core[i+1] == 'm' or Core[i+1] == None or Core[i+1] == 'd')) or Core[i] == ')':
        Core = Core[:i+1] + '@' + Core[i+1:]

Core = Core + '@'
Core = Core.split('@')
print(Core)
```

```python
NewCore = []

for i in range(0, len(Core), 1):
    Check = False
    Line = Core[i]
    
    for n in range(len(Line)):
        if Line[n] != 'x':
            Check = True
        else:
            Check = False
            break
    print(Line)
    if Line != '' and Line[-1] != ')':
        Check = False
    
            
    if Check == True:
        NewCore.append(Line)
```

```python
TrueCore = NewCore.copy()
```
