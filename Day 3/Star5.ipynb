```python
Core = open('BaseData.txt').read()
Core.replace(' ', 'x')
Count = 0
for i in range(len(Core)):
    if Core[i] != 'm' and Core[i] != 'u' and Core[i] != 'l' and Core[i] != '(' and Core[i] != ')' and Core[i] != ',' and not(Core[i].isdigit()):
        Core = Core.replace(Core[i], 'x')


for i in range(len(Core)-1):
    
    if (Core[i] == 'x' and Core[i+1] != 'm') or (Core[i] == 'm' and Core[i+1] != 'u') or (Core[i] == 'u' and Core[i+1] != 'l') or (Core[i].isdigit() and (not(Core[i+1].isdigit()) and Core[i+1] != ',' and Core[i+1] != ')')) or (Core[i] == ')' and Core[i+1] != 'm'):
        Core = Core[:i+1] + 'x' + Core[i+2:]

    if (Core[i] == 'x' and Core[i+1] == 'm') or Core[i] == ')':
        Count += 1

for i in range(len(Core) + Count - 1):
    if (Core[i] == 'x' and (Core[i+1] == 'm' or Core[i+1] == None)) or Core[i] == ')':
        Core = Core[:i+1] + '@' + Core[i+1:]

Core = Core + '@'
Core = Core.split('@')
print(Core)
        
```

```python
NewCore = []

for i in range(0, len(Core), 1):
    Check = False
    Line = Core[i]
    for n in range(len(Line)):
        if Line[n] != 'x':
            Check = True
        else:
            Check = False
            
    if Check == True:
        NewCore.append(Line)
```

```python
TrueCore = NewCore.copy()
```

```python
NewTrueCore = []

for i in range(len(TrueCore)):
    Size = False
    Line = TrueCore[i]
    Line = Line.lstrip('mul(').rstrip(')').split(',')
    for n in range(len(Line)):
        
        Line[n] = int(Line[n])
        if Line[n] > -1 and Line[n] < 1000:
            Size = True
        else:
            Size = False
        if n != 1 :
            Size = False
        elif n == 1:
            Size = True
            
    if Size == True:
        NewTrueCore.append(Line)
    TrueCore[i] = Line
print(NewTrueCore)
print(TrueCore)
```
