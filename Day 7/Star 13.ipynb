```python
AN = open('AnswersNumbers.txt').read().split('\n')
Answers = []
Numbers = []
for i in range(len(AN)):
    AN[i] = AN[i].split(':')
    Answers.append(AN[i][0])
    Numbers.append(AN[i][1])
AFile = open('Answers.txt', 'w')
AFile.write('\n'.join(Answers))
AFile.close()

NFile = open('Numbers.txt', 'w')
NFile.write('\n'.join(Numbers))
NFile.close()
```

```python
Answers = open('Answers.txt').read().split('\n')
for i in range(len(Answers)):
    Answers[i] = int(Answers[i])
Numbers = open('Numbers.txt').read().split('\n')
for i in range(len(Numbers)):
    Numbers[i] = Numbers[i].split()
    for n in range(len(Numbers[i])):
        Numbers[i][n] = int(Numbers[i][n])

total = 0
for i in range(len(Answers)):
    ThisAns = Answers[i]
    TheseNums = Numbers[i]
    bank = []
    Equal = False
    for n in range(len(TheseNums)-1):
        if n == 0:
            bank.append(TheseNums[n])
        TempNum = bank.copy()
        bank.clear()
        for m in range(len(TempNum)):
            bank.append(TempNum[m] + TheseNums[n+1])
            bank.append(TempNum[m] * TheseNums[n+1])
    if bank.count(ThisAns) != 0:
        print(ThisAns, ':', bank)
        total += ThisAns
            
print(total)       
            
        
        
        
```
