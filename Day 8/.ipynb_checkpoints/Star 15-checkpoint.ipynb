```python
Map = open('Map.txt').read().split('\n')
for i in range(len(Map)):
    Map[i] = list(Map[i])
AttenaCoor = []


for y in range(len(Map)):
    for x in range(len(Map[i])):
        if Map[y][x] != '.':
            AttenaCoor.append([Map[y][x], y, x])
print(AttenaCoor)
```
